## Задание 2

Вывести Headers данного сайта.

```python
import urllib.request

headers = urllib.request.urlopen('https://en.wikipedia.org/wiki/Python_(programming_language)').info()

print(headers)
```

## Задание 3

Вывести закодированный через base64 html код сайта, который мы взяли.

```python
import base64

html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Python_(programming_language)').read()

encoded = base64.b64encode(html)

encoded
```

## Задание 4

Используя закодированный код и модуль http.server, создайте локальный сервер, где будет выводиться заглавная страница этого сайта.

```python
from http.server import HTTPServer, BaseHTTPRequestHandler
import base64
from io import BytesIO

class SimpleHTTPRequestHandler(BaseHTTPRequestHandler):

    def do_GET(self):
        self.send_response(200)
        self.end_headers()
        self.wfile.write(base64.b64decode(encoded))

httpd = HTTPServer(('localhost', 8000), SimpleHTTPRequestHandler)
httpd.serve_forever()
```
